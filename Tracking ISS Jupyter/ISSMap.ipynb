# Final Project
This program creates a map using cartopy to show the location of the International Space Station. It also prints the longitude and latitude provided by the website from class.

```python
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
import json
import requests


#Creates the map
plt.figure(figsize = (12, 8))
ax = plt.axes(projection=ccrs.PlateCarree())
ax.coastlines()
ax.stock_img()
ax.gridlines(draw_labels=True)

#Gets ISS location from website
address = "http://api.open-notify.org/iss-now.json"
f = requests.get(address)
text = json.loads(f.text)
f.close()

#Make variables to get location from json
location = text["iss_position"]
latitude = float(location["latitude"])
longitude = float(location["longitude"])

#Plots and prints the point for ISS
ax.plot(longitude, latitude, color='blue', linewidth=3, marker='s')
print(f"Latitude: {latitude}\nLongitude: {longitude}")
plt.show()
```
